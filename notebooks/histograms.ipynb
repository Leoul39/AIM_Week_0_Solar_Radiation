```python
#Read the datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
benin_df=pd.read_csv('C:/Users/abenet/Desktop/data/benin-malanville.csv')
sierra_df=pd.read_csv('C:/Users/abenet/Desktop/data/sierraleone-bumbuna.csv')
togo_df=pd.read_csv('C:/Users/abenet/Desktop/data/togo-dapaong_qc.csv')
```

histograms for togo

```python
column=['GHI','DNI','DHI','WS','TModA','TModB']
fig,ax=plt.subplots(nrows=len(column),figsize=(8,30))
for i,j in enumerate(column):
    ax[i].hist(togo_df[j],bins=30)
    ax[i].set_xlabel(j)
plt.tight_layout()
plt.show()
```
