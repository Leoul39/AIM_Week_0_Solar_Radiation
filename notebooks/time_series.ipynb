```python
#Read the datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
benin_df=pd.read_csv('C:/Users/abenet/Desktop/data/benin-malanville.csv')
sierra_df=pd.read_csv('C:/Users/abenet/Desktop/data/sierraleone-bumbuna.csv')
togo_df=pd.read_csv('C:/Users/abenet/Desktop/data/togo-dapaong_qc.csv')
```

time series analysis for benin

```python
columns=['GHI','DNI','DHI','Tamb']
benin_df['Timestamp'] = pd.to_datetime(benin_df['Timestamp'])
fig, axes = plt.subplots(nrows=4, figsize=(10, 12))
for i, var in enumerate(columns):
    axes[i].plot(benin_df['Timestamp'], benin_df[var])
    axes[i].set_ylabel(var)

plt.xlabel('Timestamp')
plt.show()
```

time series analysis for togo

```python
columns=['GHI','DNI','DHI','Tamb']
togo_df['Timestamp'] = pd.to_datetime(togo_df['Timestamp'])
fig, axes = plt.subplots(nrows=4, figsize=(10, 12))
for i, var in enumerate(columns):
    axes[i].plot(benin_df['Timestamp'], togo_df[var])
    axes[i].set_ylabel(var)

plt.xlabel('Timestamp')
plt.show()
```
