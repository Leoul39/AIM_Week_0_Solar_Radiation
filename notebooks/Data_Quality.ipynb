```python
#Read the datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
benin_df=pd.read_csv('C:/Users/abenet/Desktop/data/benin-malanville.csv')
sierra_df=pd.read_csv('C:/Users/abenet/Desktop/data/sierraleone-bumbuna.csv')
togo_df=pd.read_csv('C:/Users/abenet/Desktop/data/togo-dapaong_qc.csv')
```

Incorrect entries for certain columns on the dataframe have been cleaned. Now, we check for outliers


outliers for togo

```python
#using z-score analysis and threshold of 3
togo_df['z_score_ws']=(togo_df['WS']-togo_df['WS'].mean())/togo_df['WS'].std()
outlier1=togo_df[abs(togo_df['z_score_ws'])>3]
togo_df['z_score_moda']=(togo_df['ModA']-togo_df['ModA'].mean())/togo_df['ModA'].std()
outlier2=togo_df[abs(togo_df['z_score_moda'])>3]
togo_df['z_score_modb']=(togo_df['ModB']-togo_df['ModB'].mean())/togo_df['ModB'].std()
outlier3=togo_df[abs(togo_df['z_score_modb'])>3]
print(len(outlier1))
print(len(outlier2))
print(len(outlier3))
```
